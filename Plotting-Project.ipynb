## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(5000)
```

```python
plt.hist(data, bins=30, color="red")
plt.title("Histogram of random 5000 normal distributed data")
plt.suptitle("Marks of Students appearing in X Board")
plt.xlabel("Marks Distribution")
plt.ylabel("# Students")
```

## BoxPlot

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(5000)
np.random.seed(10)
data1 = np.random.normal(100, 5, 500)
data2 = np.random.normal(70, 5, 500)
data3 = np.random.normal(100, 50, 1000)
data4 = np.random.normal(70, 50, 1000)
plt.boxplot([data1,data2,data3,data4])
plt.title("Boxplot of normal distributed data")
plt.suptitle("Marks of Students appearing in X Board")
plt.xlabel("Marks Distribution")
plt.ylabel("# Students")
plt.show()
```

## BarPlot

```python
import matplotlib.pyplot as plt
marks = [55,79,58,65,87]
subjects = ["Maths","Python","Data Science","SQL","Tableau"]
abc="green"
plt.bar(subjects,marks, color=abc)
for idx, val in enumerate(marks):
    plt.text(idx, val,val, ha='center')
plt.title("My Marks in Class X board")
plt.suptitle("Bar Graph")
plt.xlabel("Marks Obtained")
plt.ylabel("Subjects Appeared")
plt.show()
```

## Horizontal BarPlot

```python
import matplotlib.pyplot as plt
marks = [55,79,58,65,87]
subjects = ["Maths","Python","Data Science","SQL","Tableau"]
abc="green"
plt.barh(subjects,marks, color=abc)
for idx, val in enumerate(marks):
    plt.text(val,idx, val, ha='center')
plt.title("My Marks in Class X board")
plt.suptitle("Bar Graph")
plt.xlabel("Marks Obtained")
plt.ylabel("Subjects Appeared")
plt.show()
```

## Stacked BarPlot

```python
import matplotlib.pyplot as plt
import numpy as np
term1_marks = [55,79,58,65,77]
term2_marks = [75,89,48,69,87]
term3_marks = [73,79,58,68,97]
term4_marks = [65,69,80,67,67]

subjects = ["Maths","Python","Data Science","SQL","Tableau"]

pos = [0,1,2,3,4]
plt.bar(pos, term1_marks, color="green")
plt.bar(pos, term2_marks, bottom=term1_marks, color="yellow")
bottom_2 = np.add(term1_marks, term2_marks).tolist()
plt.bar(pos, term3_marks, bottom=bottom_2, color="blue")

bottom_3 = np.add(bottom_2, term3_marks).tolist()
plt.bar(pos, term3_marks, bottom=bottom_3, color="red")

# replacing the subject names
plt.xticks(pos,subjects, fontweight="bold")
plt.title("My Marks in Class X board")
plt.suptitle("Bar Graph")
plt.xlabel("Total Marks Obtained")
plt.ylabel("Subjects Appeared")
plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt

marks = [55,79,58,65,77]
hours = [5,8,6,7,8]

plt.scatter(hours, marks, color="red")

plt.title("My Marks in Class X board")
plt.suptitle("Scatter Graph")
plt.xlabel("Hours studied")
plt.ylabel("Total Marks Obtained")
plt.show()
```

## Pie Plot

```python
import matplotlib.pyplot as plt
import numpy

marks = [55,79,58,65,77]
subjects = ["Maths","Python","Data Science","SQL","Tableau"]
sizes = numpy.array(marks)
p,tx,autotexts = plt.pie(sizes, labels=subjects, autopct="")
for i,v in enumerate(autotexts):
    v.set_text(sizes[i])
plt.title("My Marks in Class X board")
plt.suptitle("Pie Graph")
plt.legend()
plt.show()
```
